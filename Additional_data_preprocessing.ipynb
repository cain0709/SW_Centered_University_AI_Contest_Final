```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt

def pre_resize(img):

# 이미지 읽기, 인자가 path인 경우
    if type(img) == type('string'):
        img = cv2.imread(img)

    # 1000,700 px 넘을 경우 resize
    if img.shape[1] > 1000 or img.shape[0] > 1000:
        img = cv2.resize(img, (0, 0), fx=0.5, fy=0.5, interpolation=cv2.INTER_AREA)
    elif img.shape[1] > 700 or img.shape[0] > 700:
        img = cv2.resize(img, (0, 0), fx=0.3, fy=0.3, interpolation=cv2.INTER_AREA)

    return img

def preprocessing(img): # path가 아닌 nparray

# 컬러 히스토그램 균일화
    img = pre_resize(img)
    img_ycrcb = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
    ycrcb_planes = list(cv2.split(img_ycrcb))
    ycrcb_planes[0] = cv2.equalizeHist(ycrcb_planes[0]) # 밝기 성분만 히스토그램 평활화
    dst_ycrcb = cv2.merge(ycrcb_planes)
    img_result = cv2.cvtColor(dst_ycrcb, cv2.COLOR_YCrCb2BGR)

    return img_result
```

```python
def pre_resize(img):

    # 이미지 읽기, 인자가 path인 경우
    if type(img) == type('string'):
        img = cv2.imread(img)

    # 1000,700 px 넘을 경우 resize
    if img.shape[1] > 1000 or img.shape[0] > 1000:
        img = cv2.resize(img, (0, 0), fx=0.5, fy=0.5, interpolation=cv2.INTER_AREA)
    elif img.shape[1] > 700 or img.shape[0] > 700:
        img = cv2.resize(img, (0, 0), fx=0.3, fy=0.3, interpolation=cv2.INTER_AREA)

    return img
```

```python
from sklearn.cluster import MeanShift, estimate_bandwidth

# in (np.ndarray | path string ) / out (np.ndarray)
def pre_meanshift(img,q=3, n=50):

    # 이미지 읽기, 인자가 path인 경우
    if type(img) == type('string'):
        img = cv2.imread(img)

    # medianBlur 처리, k=3
    img = cv2.medianBlur(img, 3)

    # flatten the image
    flat_image = img.reshape((-1,3))
    flat_image = np.float32(flat_image)

    # meanshift
    # - quantile : 페어 별 거리의 분위수, 0.5일 때 중앙값이 사용됨 (디폴트 0.3) -> 낮을 수록 매우 비슷해야 클러스터링됨. 세그먼트 갯수가 많아짐
    # - n_samples : 사용할 샘플 수 : 디폴트 : 모든 샘플 사용 -> 높을 수록 유사하면 다묶임. 세그먼트 갯수 적어짐
    # - max_iter: 클러스터링 수립x일 때 반복 횟수 (디폴트 300) 많아야 잘 묶음
    bandwidth = estimate_bandwidth(flat_image, quantile=q, n_samples=n)
    ms = MeanShift(bandwidth=bandwidth, max_iter=800, bin_seeding=True)
    ms.fit(flat_image)
    labeled=ms.labels_

    # get number of segments
    segments = np.unique(labeled)
    print('Number of segments: ', segments.shape[0])

    # get the average color of each segment
    total = np.zeros((segments.shape[0], 3), dtype=float)
    count = np.zeros(total.shape, dtype=float)
    for i, label in enumerate(labeled):
        total[label] = total[label] + flat_image[i]
        count[label] += 1
    avg = total/count
    avg = np.uint8(avg)

    # cast the labeled image into the corresponding average color
    res = avg[labeled]
    result = res.reshape((img.shape))

    return result

    # using function
    # input = train_df['img_path'][905]]]]
```
